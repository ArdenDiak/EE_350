### 1.2 Domaine des fréquences

Le signal ci-dessus oscille à une fréquence de 10Hz. Dans le domaine temporel, cela est illustré par une "onde" qui fait une boucle entre les mêmes valeurs dans le temps. Le domaine temporel est très inefficace pour traiter les signaux oscillants. Heureusement, le domaine fréquentiel est plus pertinent pour ce type de signaux. Ces domaines sont liés par la transformée de Fourier. Pour prouver cette affirmation, traçons la transformée de Fourier de notre signal. 

Encore une fois, utilisez `NumPy` pour calculer la transformée de Fourier de votre signal définie ci-dessus. Traditionnellement, la transformée de Fourier est représentée des fréquences négatives aux fréquences positives. NumPy produit une transformée de Fourier pour les fréquences positives. Vous pouvez passer à la représentation habituelle avec la fonction `np.fft.fftshit()`.

Une fois que vous avez calculé la transformée de Fourier, tracez sa valeur absolue, sa partie réelle et sa partie imaginaire.


````python
freq= 10.0 #Hz
sampFreq= 100.0 #Hz
numCycles= 4.0

t= np.arange(0, numCycles * (1/freq), 1/sampFreq)
x= np.cos(2*np.pi*freq*t)

f= np.arange(-sampFreq/2, sampFreq/2, 10/4)
X_f= np.fft.fft(x);
X_f_Re= np.real(X_f);
X_f_shifted= np.fft.fftshift(X_f)

fig, axs = plt.subplots(3,1)
fig.tight_layout()

axs[0].plot(f, X_f_shifted.real)
#axs[0].set_xlabel("Frequency [Hz]")
axs[0].set_ylabel("Magnitude")
axs[0].set_title("TF: Valeur absolue")

axs[1].plot(f, abs(X_f_shifted.real))
#axs[1].set_xlabel("Frequency [Hz]")
axs[1].set_ylabel("Magnitude")
axs[1].set_title("TF: Partie réelle")

axs[2].plot(f, X_f_shifted.imag)
axs[2].set_xlabel("Frequency [Hz]")
axs[2].set_ylabel("Magnitude")
axs[2].set_title("TF: Partie imaginaire")

plt.show()
````

# 3 Approximation de la fonction

La base $ \mathcal{S} \in \mathbb{R}^{M \times N} $ et le vecteur temps discret $ n \in \mathbb{R}^N $ sont générés en utilisant la fonction `build_basis()` définie ci-dessous.

Chaque signal $s_i(n)$ représente la ligne $i^{th}$ de $\mathcal{S}$ et tous les $s_i(n)$, $i \in \{1,2,...,M\}$ sont linéairement indépendants.

````python
def build_basis(degree, n):
    """
    Build basis
    :param degree: Degree of the polynome
    :param n:      Number of points to sample
    """
    nt = np.arange(-1.0, 1.0, 2 / n)
    basis = np.zeros(shape=(degree, n))
    for k in range(degree):
        basis[k, :] = nt ** k
    return basis, nt

# Basis + Time
S, n = build_basis(degree=35, n=2000)
````

## Orthogonalité

Vérifiez si les bases en $S$ sont orthogonales. Implémentez la fonction `orthocheck(...)`, qui prend en entrée une matrice $U$ stockant la base sur chaque ligne et donnez en sortie `True` si $u_i\left(n\right)$, $i \in {1, 2, . . . M}$ forment une base orthogonale, sinon "Faux".

*Conseil : Lorsque vous implémentez la fonction, les valeurs inférieures à 1e-4 peuvent être considérées comme des zéros*

````python
def orthocheck(basis, tol=1e-4):
    """
    Check if a given basis is orthogonal.
    :param basis: Numpy array storing base vector on every row.
    :return: `True` if the basis is orthogonal, `False` otherwise.
    """
    M= basis.shape[0]
    N= basis.shape[1]
    
    for i in range(M):
        for j in range(M):
            
            if(i != j):
                result= np.dot(basis[i,:], basis[j,:])
                
                if(abs(result) >= tol): #not orthogonal
                    #print("basis[{},:] and basis[{},:] are not orthogonal".format(i,j))
                    return False;
    
    return True;
````

````python
# Is basis orthogonal
if orthocheck(S):
    print('Les signaux d\'entrée forment un ensemble orthonormé !')
else:
    print('Les signaux d\'entrée ne forment PAS un ensemble orthonormé !')
````

## 3.3 Gram-Schmidt

Implémente la fonction `modified_gramschmidt(...)`, qui prend la matrice $U$ en entrée et applique l'algorithme de Gram-Schmidt aux signaux $u_i\left(n\right)$, $i \in {1,2,...,M}$ et sort la nouvelle base dans la matrice $B$.

Il s'avère que la procédure de Gram-Schmidt que nous avons introduite dans le cours souffre d'instabilité numérique : Les erreurs d'arrondi peuvent s'accumuler et détruire l'orthogonalité des vecteurs résultants. C'est pourquoi une procédure Gram-Schmidt modifiée est introduite pour aider à remédier à ce problème.

```
def modified_gramschmidt(U):
for j = 1:n
    vj=xj
    
for  j=1:n
    qj= vj / ‖vj‖2
      
    for k = j+1:n
        vk = vk − (qj' vk) qj
        
return v
```

Vérifiez que la nouvelle base $\phi_i\left(n\right)$, $i \in {1, 2, . . . M }$ sont bien orthonormés. Tracez les signaux $\phi_i\left(n\right)$ pour $i \in {1, 2, . . . 8}$ dans le même graphe et commentez les signaux résultants générés par l'algorithme de Gram-Schmidt.

````python
def modified_gramschmidt(basis):
    """
    Apply Gram-Schmidt to orthogonalize a given basis.
    :param basis: Numpy array storing base vector on every row.
    :return: Numpy array with orthogonalized basis.
    """
    M= basis.shape[0]
    N= basis.shape[1]
    v= np.zeros(basis.shape);
    q= np.zeros(basis.shape[1]);
    
    for j in range(0,M):
        v[j,:]= basis[j,:]
    
    for j in range(0,M):
        q= v[j,:] / np.sqrt(np.dot(v[j,:],v[j,:]))
        
        for k in range(j+1,M):
            v[k,:] = v[k,:] - (np.dot(q, v[k,:]) * q)
    
    return v;
````

````python
# Apply Gram-Schmidt 
Sn = modified_gramschmidt(S)

if orthocheck(Sn):
    print('Les signaux d\'entrée forment un ensemble orthonormé !')
else:
    print('Les signaux d\'entrée ne forment PAS un ensemble orthonormé !')

# Solution
plot_basis(Sn[:30, :], n)
````

The amplitude of the basis' functions have been rescaled by a constant, allowing the original functions to form an orthonormal basis. 

## 3.4 Approximation

Considérons le signal $f(t)$ défini comme :
$$
f\left(t\right) = \left(-2\left| t \right| + 1 \right)^3 + \frac{1}{2} \left(-2\left| t \right| + 1 \right)^2 
$$


Définissez $f(n)$ en échantillonnant le signal $f(t)$ à l'aide du vecteur $n$. Écrire un script pour projeter $f(n)$ sur la base $\phi_i\left(n\right)$, $i \in {1,2,...,K}$ avec $K \in {5, 10, 15, 20, 25, 30, 35}$. Tracez le signal $f(n)$ et ses approximations $\hat{f}_K\left(n\right)$ pour $K \in {5, 10, 15, 20, 25, 30, 35}$ et commentez ce que vous obtenez.

L'erreur de reconstruction entre $f(n)$ et $\hat{f}_K$ est mesurée à l'aide de l'erreur quadratique moyenne (*MSE*) et est définie comme
$$
MSE\left(K\right) = \frac{1}{N}\left|\left|f(n) -  \hat{f}_K\right|\right|^2
$$

Tracez l'évolution de l'erreur quadratique moyenne par rapport à $K$ et commentez ce que vous observez.

````python
def approximate(K, N, basis):
    M= basis.shape[0]
    
    nt= np.arange(-1.0,1.0,2/N)
    f_n= (((-2 * abs(nt)) + 1) ** 3) + (0.5 * (( (-2 * abs(nt)) + 1) ** 2)) 
    v= f_n
    
    proj_S_v= np.zeros(nt.shape)
    for i in range(0,K):
        u_i= basis[i,:]
        proj_S_v+= (np.dot(u_i, v) / np.dot(u_i, u_i)) * u_i
    
    mse= (1/N) * np.dot(f_n - proj_S_v, f_n - proj_S_v)
    return mse

n= list()
Y=list()
for i in range(5,40,5):
    mse= approximate(i,2000,Sn)
    n.append(i)
    Y.append(mse)
    print("{}: {}".format(i,mse))
    
fig, axs= plt.subplots(1)
axs.plot(n,Y)
axs.set_xlabel("K")
axs.set_ylabel("MSE(K)")
axs.set_title("MSE from approximating f(t) with basis vectors phi_i(n)")
plt.show()
    
````

- The more basis vectors are used to approximate the sampled function $f(n)$, the lower the mean squared error $MSE(k)$. A linear combination of a greater number of basis polynomials will more closely approximate the function $f(n)$.
