## 1 Transformée de Fourier discrète - TFD

La TFD $X\left[k\right]$ d'un signal $x\left[n \right]$ de longueur $N$ est donnée par:
$$
X\left[k\right] = \sum_{n = n_0}^{n_0 + N - 1} x\left[n \right] W_{N}^{-kn}
$$

et sa transformée inverse est donnée par:
$$
x\left[n \right] = \frac{1}{N} \sum_{k = 0}^{N - 1} X\left[k\right]  W_{N}^{kn}
$$
où $W_{N} = \text{exp}j\frac{2\pi}{N}$.

#### 1.1 Base de la TFD

L'équation ci-dessus décrit la projection d'un signal $x$ sur une base. Identifiez les fonctions de cette base.



The basis functions are $W_N^{-kn}=\exp(-j\frac{2\pi}{N}kn)$. 
The following equation for calculating the DFT $X[k]$ describes the projection of $x[n]$ onto that basis $W_N^{-kn}$ where $n\in[n_0, n_0+N-1]$

$$
X[k]= \sum_{n=n_0}^{n_0+N-1} x\left[n\right]W_N^{-kn}
$$ 

#### 1.2 TFD d'un vecteur de la base

Le signal $x_w[n]$ est défini comme $x_w[n] = w[n] \cdot x[n]$, où $w[n]$ est une fenêtre rectangulaire de taille $N=100$ et $x[n]=\cos \left( \frac{2 \pi f n}{f_s} \right)$, avec $f = 90$ Hz et $f_s = 1000$ Hz.

```python
def dB(x):
    return 20 * np.log10(np.abs(x))

def dft_xn(N, window_type, ax1, ax2, color):
    #sample spacing
    
    numCycles= 1.0
    f_s= 1000 #Hz
    T=1/f_s
    
    n= np.arange(N)
    x= np.cos((2*np.pi*90*n)/f_s)

    w= spsig.get_window(window_type,N)
    yw= x*w
    
    ywf= np.fft.fft(yw)
    f= np.linspace(-f_s/2, f_s/2, len(ywf))

    
    ax1.stem(n, yw, color[0])
    ax2.stem(f, np.abs(ywf), color[1])

    ax1.set_xlabel("$n$")
    ax1.set_ylabel("Amplitude")
    ax2.set_xlabel("Frequency [Hz]")
    ax2.set_ylabel("Amplitude")
    
    ax1.set_title("$x_w[n]$ N={} points (${}$)".format(N, window_type))
    ax2.set_title("DFT of $x_w[n]$ N={} points (${}$)".format(N, window_type))
    
    #count non-zero points in DFT
    ct=0
    for ele in np.abs(ywf):
        if ele > 10e-2:
            ct+= 1
    print("DFT of x_w[n] with {} window of size {} has {} positive points ".format(window_type, N, ct))
    return ct

    
fig,axs= plt.subplots(1,2, figsize=(16,4))
fig.subplots_adjust(hspace = 0.4)

dft_xn(100, 'boxcar',axs[0], axs[1], ['b','r'])

plt.show()
```

#### 2.1. Réponse impulsionelle et fonction de transfert

Considérons un système linéaire et invariant dans en position $h$ qui prends en entrée une image en nuances de gris et produit une nouvelle image en nuance de gris $v$, comme illustré sur la figure 1.

</p>

L'entrée et la sortie sont reliées par l'opération de convolution, notée $*$. La sortie corresponds à la convolution de $y$ par la réponse impulsionelle du système $h$, soit $v = y * h$. En 2D la convolution est définie pour chaque point $(n, m)$ comme: 

$
\begin{aligned}
    v\left(n, m\right)  & = \left(y * h\right)\left( n, m \right) \\
                        & = \sum_{o=-\infty}^{+\infty}\sum_{p=-\infty}^{+\infty} h\left(o, p\right) y\left(n-o, m-p\right).
\end{aligned}
$

Cette équation exprime chaque pixel $(n, m)$ de l'image de sortie comme une combinaison linéaire de pixels de l'entrée. La transformée de Fourier 2D est reliée à la convolution par une propriété importante:

$
\begin{aligned}
V\left(k, l\right) & = \mathcal{F}\left(v\right)\\
                & = \mathcal{F}\left(y * h \right) \\
                & = \mathcal{F}\left(y\right)\mathcal{F}\left(h\right) \\
                & =   Y\left(k, l\right)H\left(k, l\right).
\end{aligned}
$

où $k, l$ sont les fréquences 2D, $\mathcal{F}$ est la transformée de Fourier et $V\left(k, l\right), Y\left(k, l\right), H\left(k, l\right)$ sont les transformées de Fourier de $v\ , y\ , h$ repectivement.

Dans le domaine de Fourier la convolution se résume à une simple multiplication des transformées de Fourier  $Y(k, l)$ et $H(k, l)$. La fonction $H(k, l)$ est définie comme la fonction de transfert du système $h$. Dans cet exercice, nous allons montrer que le calcul de $v\left(n, m\right)$ avec ces deux méthodes est équivalent (quand il est fait correctement).


#### 2.2. Domaine spatial

Définissons $h(n,m)$. C'est un filtre gaussien de variance 5 et de taille $(N_1, N_2) = (21, 21)$. Écrivez une fonction `gkernel` qui prends en entrée un taille $N=N_1 =N_2$ de filtre et sa variance et qui renvoie la réponse impulsionelle 2D `h`. Affichez le filtre grâce au module  `matplotlib.pyplot` importé ci-dessus.

```python
def gkernel(N, sigma):
    delta= spsig.unit_impulse((N,N),'mid')
    h= scipy.ndimage.filters.gaussian_filter(delta, sigma)
    return h

h= gkernel(21,5)
plt.imshow(h, cmap='gray')
plt.show()
```

En utilisant la définition de la convolution, quelle est l'expression mathématique de la convolution avec le kernel ? Plus précisément, quelles sont les bornes des sommes ? Pourquoi ?

For an image of size NxM, the limits of the convolution sum are N and M, the size of the image. Given input image $y(n,m)$ and gaussian kernel (2D impulse response) $h(n,m)$, we have that the output of convoluting an input Black/White image with the gaussian kernel is:

$
\begin{aligned}
 \left(y * h\right)\left( n, m \right) & = \sum_{o=0}^{N}\sum_{p=0}^{M} h\left(o, p\right) y\left(n-o, m-p\right).\\
\end{aligned}
$
