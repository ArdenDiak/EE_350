# 1. Fonction de transfert (5 pts)


On considère la fonction de transfert:

$$
H(z) = \frac{0.2066+0.4132z^{-1}+0.2066z^{-2}}{1-0.3695z^{-1}+0.1958z^{-2}}
$$

### 1.1 Filtre, pôles et zéros
Trouver les pôles et les zéros de $H(z)$. Vous pouvez vous aider de la fonction [tf2zpk](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.tf2zpk.html). Quel est le gain de ce filtre ?
Le gain de ce filtre est 0.2066

```python
b= np.array([0.2066, 0.4132, 0.2066])
a= np.array([1, -0.3695, 0.1958])

(z, p, k)= tf2zpk(b,a)

print("Zeros: {}".format(z))
print("Poles: {}".format(p))
print("Gain: {}".format(k))
```

Dessiner les pôles et les zéros  de $H(z)$ dans le plan complexe.

```python
fig, ax1 = plt.subplots(1)
ax1.set_xlabel('Real')
ax1.set_ylabel('Imaginary')
ax1.set_title('Pôles et les zéros de H(z)')

ax1.scatter(p.real,p.imag)
ax1.scatter(z.real,z.imag)
plt.show()
```

# 2. Système causal (4 pts)

Soit une famille de systèmes parametrée par $N$ avec transformées en z suivantes:

$$ H_N(z) = (z^{-1} - 2)^N - z^{-N} $$

1. Donnez une formule pour les réponses impulsionelles de ces systèmes.

$$
    H_N(z) = \sum_{k=0}^{n-1} {N \choose k} (-2)^{N-k} * (z^{-1})^{k} 
$$
    Take the inverse z-transform of both sides, using the table of z-transforms to simplify the expression
$$
    h_N(n) = \sum_{k=0}^{n-1} {N \choose k} (-2)^{N-k} * \delta[n-k]
$$

2. Calculez, numériquement pour $N \in \{2, 4, 6, 10\}$ les zéros de ces systèmes et affichez les sur un diagramme pôles-zéros. Vous pouvez vous aider de la fonction `scipy.special.binom`.

```python
def getPolesZeros(N):
    numer= list()
    for i in range(0,N):
        ele= binom(N, i) * ((-2) ** (N-i))
        numer.append(ele)

    return (numer,1) #denominator is 1



def graphCircle(ax):
    x= np.linspace(-1,1,100)
    y= np.sqrt(1 - (x ** 2))
    
    ax.plot(x,y, color='b')
    ax.plot(x,-y, color='b')
    

fig, axs = plt.subplots(2,2)
fig.tight_layout()
axs[1,0].set_xlabel('Real')
axs[1,1].set_xlabel('Real')
axs[0,0].set_ylabel('Imaginary')
axs[1,0].set_ylabel('Imaginary')


#plot the poles and zeros for each N
test_set= [2,4,6,10]
graph_set= [(0,0),(0,1),(1,0),(1,1)]
for i in range(4):
    (b,a)= getPolesZeros(test_set[i])
    (z,p,k)= tf2zpk(b, a)

    ax= axs[graph_set[i][0],graph_set[i][1]]    
    ax.scatter(z.real,z.imag)
    graphCircle(ax)
    ax.set_title("Zéros pour N={}".format(test_set[i]))
    
    
plt.show()
```

3. Sur la base de la position des zéros uniquement, quel est le type de cette famille de filtres (passe-haut, passe-bas, passe-bande, coupe-bande)? 
    * Based on the graph of the zeros of the transfer function, this family of filters describes a high-pass filter, as the poles are located only in the first and fourth quardants of the graph, and since there are zeros near the frequency $\omega=0$

4. Affichez leurs réponses en fréquence (en amplitude et en phase, avec une échelle logarithmique pour le premier) pour les valeurs $N \in \{2, 4, 6, 10\}$. Quel est l'effet de l'augmentation de $N$? (Avant d'afficher la réponse en amplitude des filtres, normaliser leur gain, i.e. divisez les coeficients de la réponse impulsionelle par la somme de leurs valeurs absolues)
    - As $N$ increases, the filter $H_N(z)$ attenuates lower frequencies much more. In other words the filter's gain increases, since there are more poles are located in quadrants I and IV of the graph (as seen in the pole-zero diagrams in the previous question). Frequencies close to 0 rad/sample are most attenuated by this filter, and the phase shift decreases faster with greater even values of $N$

```python
def plotMagPhase(b,a,ax1,ax2,C,legend_in):
    w, h= freqz(b,1,1024)

    ax1.plot(w, 20 * np.log10(abs(h)), color=C)
    ax1.legend(legend_in)
    
    angles_rad = np.unwrap(np.angle(h))
    ax2.plot(w, angles_rad, 'g', color=C)
    ax2.set_ylabel('Angle [rad]')

    ax2.legend(legend_in)
    
    
fig, axs = plt.subplots(2,1)
fig.subplots_adjust(hspace = 0.4)
axs[0].set_title("Frequency Response")
axs[0].set_xlabel('Frequency [rad/samples]')
axs[0].set_ylabel('Amplitude [dB]')

axs[1].set_title("Phase Diagram")
axs[1].set_xlabel('Frequency [rad/samples]')
axs[1].set_ylabel('Amplitude [dB]')



#plot the amplitude and phase of each TF's freq response
legend_in=["N=2", "N=4", "N=6", "N=10"]
test_set= [2,4,6,10]
color_set=['b','r','y','g']
for i in range(4):
    (b,a)= getPolesZeros(test_set[i])
    
    #normalize each TF's gain
    norm= 0
    for ele in b:
        norm += abs(ele)
    b /= norm
    
    ax= graph_set[i]
    plotMagPhase(b,a,axs[0],axs[1],color_set[i],legend_in)

    
plt.show()
```

5. Répétez les points 2.,3. et 4. pour les systèmes du type:
$$ \hat{h}_N[n] = \begin{cases} h_N[n/2] \text{ si } n \text{ pair} \\ 0 \text{ sinon}\end{cases}$$
où $h_N$ est la réponse impulsionelle des filtres définis ci-dessus. Quelle est l'effet de la transformation $h_N \implies \hat{h}_N$ ?

* Looking at the entry in the z-transform table for the Z-transform of a time expansion by some integer k, the effect of the transformation $h_N \implies \hat{h}_N$ is that the transfer function $H_N(z)$ becomes
$$ H_N(z) = \sum\limits_{k=0}^{N-1} \binom{N}{k} (-2)^{N-k} * (z^{-2k})$$

* Furthermore, Since $N\in{2,4,6,10}$ could only take even values when considering $h_N[n]$, the transformation $h_N \implies \hat{h}_N$ introduces odd values of N, which alters the behavior of the filter's frequency response (as observed in its amplitude and phase plots).

* $\textbf{Question 2.3}$: Sur la base de la position des zéros uniquement, quel est le type de cette famille de filtres (passe-haut, passe-bas, passe-bande, coupe-bande)? 
    * Based on the graph of the zeros of the transfer function, this family of filters describes a bandpass filter (from observing the amplitude graph of the transfer function).

* $\textbf{Answer to 2.4}$: As $N$ increases, the gain of the filter increases, and the slope of the phase shift in the phase diagram increases (the phase is changing more rapidly).

```python
def getPolesZeros_new(N):
    numer = np.zeros(N*2-1)
    for i in range(0,N):
        numer[2*i] = binom(N,i)*((-2)**(N-i))

    return (numer,1) #denominator is 1

# Question 2.2 
fig, axs = plt.subplots(2,2)
fig.subplots_adjust(hspace = 0.4)

axs[1,0].set_xlabel('Real')
axs[1,1].set_xlabel('Real')
axs[0,0].set_ylabel('Imaginary')
axs[1,0].set_ylabel('Imaginary')


#plot the poles and zeros for each N
test_set= [2,4,6,10]
graph_set= [(0,0),(0,1),(1,0),(1,1)]
for i in range(len(graph_set)):
    N= int(test_set[i]/2) #since all N in the given set are even
    (b,a)= getPolesZeros(N) 
    (z,p,k)= tf2zpk(b, a)

    ax= axs[graph_set[i][0],graph_set[i][1]]    
    graphCircle(ax)
    ax.scatter(z.real,z.imag)
    ax.set_title("Zéros pour N={}".format(N))


#Question 2.3: see above
#Question 2.4
fig, axs = plt.subplots(2,1)
fig.subplots_adjust(hspace = 0.4)
axs[0].set_title("Frequency Response")
axs[0].set_xlabel('Frequency [rad/samples]')
axs[0].set_ylabel('Amplitude [dB]')

axs[1].set_title("Phase Diagram")
axs[1].set_xlabel('Frequency [rad/samples]')
axs[1].set_ylabel('Amplitude [dB]')

#plot the amplitude and phase of each TF's freq response
test_set= [2,4,6,10]
color_set=['b','r','y','g']
legend_in=["N=1","N=2","N=3","N=5"]
for i in range(4):
    (b,a)= getPolesZeros_new(int(test_set[i]/2))
    
    #normalize each TF's gain
    norm= 0
    for ele in b:
        norm += abs(ele)
    b /= norm
    
    ax= graph_set[i]
    plotMagPhase(b,a,axs[0],axs[1],color_set[i],legend_in)

    
plt.show()
```

# 3. Stabilité (3 pts)

Soit les systèmes suivant:
$$ G_1(z) = \frac{1 + 0.3 z^{-1}}{1 + 0.8z^{-1} - 0.3 z^{-2}}$$
$$ G_2(z) = \frac{1 + 0.3 z^{-1}}{1 + 0.5z^{-1} - 0.3 z^{-2}}$$
1. Pour les deux systèmes, affichez les 32 premiers éléments de leurs réponses impulsionelles ainsi que leur pôles et leurs zéros sur le diagramme du même nom. Est-ce que les systèmes sont stables?  
    - $G_1(z)$ doesn't describe a stable system because it is not absolutely summable in the frequency domain (the graph its impulse response shows divergence). $G_2(z)$ describes a stable system because it is absolutely summable in the frequency domain (the graph its impulse response shows convergence to 0).

```python
#G_1(z)
b1= np.array([1, 0.3])
a1= np.array([1, 0.8, -0.3])

#G_2(z)
b2= np.array([1, 0.3])
a2= np.array([1, 0.5, -0.3])

#Poles and Zeros
(z1, p1, k1)= tf2zpk(b1,a1)
(z2, p2, k2)= tf2zpk(b2,a2)

fig_1, axs= plt.subplots(1,2,figsize=(10,5))
fig_1.subplots_adjust(hspace = 0.4)

axs[0].set_title('G_1 Poles and Zeros')
axs[0].set_xlabel('Real')
axs[0].set_ylabel('Imaginary')
axs[0].scatter(z1.real, z1.imag)
axs[0].scatter(p1.real, p1.imag)
graphCircle(axs[0])

axs[1].set_title('G_2 Poles and Zeros')
axs[1].set_xlabel('Real')
axs[1].scatter(z2.real, z2.imag)
axs[1].scatter(p2.real, p2.imag)
graphCircle(axs[1])

#Impulse Response
tout1, yout1= dimpulse((b1,a1,1), n=32)
tout2, yout2= dimpulse((b2,a2,1), n=32)

fig_2, axs_2= plt.subplots(1,2,figsize=(12,4))
fig_2.subplots_adjust(hspace = 0.4)

axs_2[0].set_title('G_1 Impulse Response')
axs_2[0].set_xlabel('n')
axs_2[0].set_ylabel('n')
axs_2[0].stem(tout1, np.squeeze(yout1))
axs_2[1].set_title('G_2 Impulse Response')
axs_2[1].set_xlabel('n')
axs_2[1].stem(tout2, np.squeeze(yout2))


plt.show()
```

2.Supposons qu'en fait le système $G_\alpha(z)$ soit implémenté avec une puce électronique défectueuse qui présente une valeur de $\alpha$ uniformément distribuée entre $0$ et $1$ à chaque fois que le système est utilisé.
$$ G_\alpha(z) = \frac{1 + 0.3 z^{-1}}{1 + \alpha z^{-1} - 0.3 z^{-2}}$$
Estimez numériquement la probabilité que le système $G_a(z)$ soit stable. Vous pouvez vous aider de la fonction `np.random.rand()`.



```python
#G_a(z)
alpha= np.random.rand()

b1= np.array([1, 0.3])
a1= np.array([1, alpha, -0.3])

N=1000
#Poles and Zeros
P= list()
for i in range(0,N):
    alpha= np.random.rand()
    a1= np.array([1, alpha, -0.3])
    (z1, p1, k1)= tf2zpk(b1,a1)
    P.append(max(abs(p1))<1)

print("The system $G_a(z)$ has a probability P={}% of being stable (based on {} samples)\n".format((sum(P)/N)*100, N))
```
